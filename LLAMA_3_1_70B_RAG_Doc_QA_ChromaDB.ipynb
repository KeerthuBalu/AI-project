```python
# laoding the document
loader = UnstructuredFileLoader("attention_is_all_you_need.pdf")
documents = loader.load()
```

```python
# splitting into text chunks
text_splitter = CharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=400
)
```

```python
texts = text_splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory = "doc_db"
```

```python
vectordb = Chroma.from_documents(
    documents=texts,
    embedding=embeddings,
    persist_directory=persist_directory
)
```

```python
# retriever
retriever = vectordb.as_retriever()
```

```python
# llm from groq
llm = ChatGroq(
    model="llama-3.1-70b-versatile",
    temperature=0
)
```

```python
# create a qa chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
# invoke the qa chain and get a response for user query
query = "What is the major keypoints discussed in this paper?"
response = qa_chain.invoke({"query": query})
```
